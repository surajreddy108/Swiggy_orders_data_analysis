**Restaurant ranked based on no of orders**

```python
restaurant_orders=orders.groupby('restaurant_name')['order_id'].nunique().reset_index()
restaurant_orders['rank']=restaurant_orders['order_id'].rank(method='first',ascending=False)
restaurant_orders.sort_values(by=['order_id'],ascending=False)
#restaurant_orders
```

**Monthwise orders**

```python
orders['order_time']=orders['order_time'].apply(pd.to_datetime)
order_month=orders['order_time'].dt.month.reset_index()
order_month
order_month_freq=order_month.groupby('order_time')['index'].count()

order_month_freq.reset_index()
#orders['order_time'].dt.strftime('%m').reset_index().groupby('order_time')['index'].count().reset_index()
```

**Latest Order**

```python
orders['order_time']=orders['order_time'].apply(pd.to_datetime)
#orders['order_time'].dt.strftime('%m')
orders['order_time'].max()
```

**Total amount spent on orders monthwise**

```python
orders
order_value_monthwise=orders.groupby(orders['order_time'].dt.strftime('%m'))['order_total'].sum().reset_index()
order_value_monthwise
order_value_monthwise.sort_values(by='order_total',ascending=False)
```

**Amount spent in each restaurant**

```python
restaurant_spent=orders.groupby('restaurant_name')['order_total'].sum().reset_index()
restaurant_spent=restaurant_spent.sort_values(by='order_total',ascending=False)
restaurant_spent
```

**Daywise orders submitted**

```python
order_weekdaywise=orders['order_time'].apply(pd.to_datetime).dt.strftime('%w').reset_index()
order_weekdaywise
order_weekdaywise.groupby('order_time')['index'].count()
```

**Yearwise amount spent on orders with ranks**

```python
orders['Year']=orders['order_time'].apply(pd.to_datetime).dt.strftime('%Y')
orders
orders_yearwise=orders.groupby('Year')['order_total'].sum().reset_index()
orders_yearwise_sorted=orders_yearwise.sort_values(by='order_total', ascending=False)
orders_yearwise_sorted['rank']=orders_yearwise_sorted['order_total'].rank(method='first',ascending=False)
orders_yearwise_sorted
```
